#### Prioritized Sorting of a Spark Data Frame

```
courses_df.show()
```

* Make sure the data is sorted in custom order by level and then in numerially in descending order by number of ratings.
* All the beginner level courses should come first, followed by intermediate level and then by Advanced level.

```
from pyspark.sql.functions import when,lit
```

```
courses_level = when(col('suitable_for')=='Beginner',0).\
                when(col('suitable_for')=='Intermediate',1).\
                otherwise(2)
```

```
courses_df.withColumn('courses_level',lit(courses_level)).show()
```

```
courses_df.\
  orderBy(courses_level,col('number_of_ratings').desc()).show()
```

```
# Using expr and then case and when

from pyspark.sql.functions import expr

course_level = expr("""
     case
     when suitable_for = 'Beginner' then 0
     when suitable_for = 'Intermediate' then 1
     else 2
     end
         """)

courses_df.\
  sort(course_level,col('number_of_ratings').desc()).show()
```
