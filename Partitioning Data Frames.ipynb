* Partition data by date

```
from pyspark.sql.functions import date_format
```

```
orders.\
  withColumn('order_date',date_format('order_date','yyyyMMdd')).show()
```

```
orders.\
  withColumn('order_date', date_format('order_date','yyyyMMdd')).\
  coalesce(1).\
  write.\
  partitionBy('order_date').\
  parquet(f'/user/{username}/retail_db/orders_partitioned_by_date')
  
```

* Partition data by month

```
orders.\
  withColumn('order_month',date_format('order_date','yyyyMM')).\
  coalesce(1).\
  write.\
mode('overwrite').\
parquet(f'/user/{username}/retail_db/orders_partitioned_by_month',partitionBy='order_month')
```

```
dbutils.fs.ls(f'/user/{username}/retail_db/orders_partitioned_by_month')
```

```
spark.read.parquet(f'/user/{username}/retail_db/orders_partitioned_by_month').count()
```

```
spark.read.parquet(f'/user/{username}/retail_db/orders_partitioned_by_month').show()
```

* Partition data by year, month and then day_of_month

```
orders.\
  withColumn('year',date_format('order_date','yyyy')).\
  withColumn('month',date_format('order_date','MM')).\
  withColumn('day_of_month',date_format('order_date','dd')).\
  coalesce(1).\
  write.\
  partitionBy('year','month','day_of_month').\
  parquet(f'/user/{username}/retail_db/orders_partitioned')
  
```

```
dbutils.fs.ls(f'/user/{username}/retail_db/orders_partitioned')
```

```
dbutils.fs.ls(f'/user/{username}/retail_db/orders_partitioned/year=2014/month=01/day_of_month=09')
```
